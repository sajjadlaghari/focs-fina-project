### 1. Create a single dataframe with the concatenation of all input csv files, adding a column called country

```python
import pandas as pd
from pathlib import Path

files = ['CAvideos.csv','DEvideos.csv','FRvideos.csv','GBvideos.csv','INvideos.csv','JPvideos.csv']
dfs = []

for f in files:
    country = Path(f).stem.replace('videos','').upper()
    try:
        df = pd.read_csv(f, encoding='utf-8')
    except UnicodeDecodeError:
        df = pd.read_csv(f, encoding='ISO-8859-1')
    df['country'] = country
    dfs.append(df)

merged = pd.concat(dfs, ignore_index=True, sort=False)
print("Step 1: Combined shape:", merged.shape)
```

### 2. Extract all videos that have no tag.

```python
no_tag_videos = merged[merged['tags'].isna() | (merged['tags'].str.strip() == '') | (merged['tags'].str.strip() == '[]')]
print("Step 2: Videos with no tags:", len(no_tag_videos))
```

### 3. For each channel, determine the total number of views


```python
merged['views'] = pd.to_numeric(merged['views'], errors='coerce').fillna(0)
channel_views = merged.groupby('channel_title')['views'].sum().sort_values(ascending=False).reset_index()
print("Step 3: Top channels by total views\n", channel_views.head(10))
```

###  4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded, and remove those rows from the original dataframe.


```python
for col in ['comments_disabled','ratings_disabled','video_error_or_removed']:
    merged[col+'_bool'] = merged[col].astype(str).str.lower() == 'true'

excluded = merged[(merged['comments_disabled_bool']) | 
                  (merged['ratings_disabled_bool']) | 
                  (merged['video_error_or_removed_bool'])].copy()

cleaned = merged.drop(excluded.index)
print("Step 4: Excluded rows:", len(excluded), "Cleaned shape:", cleaned.shape)
```

### 5. Add a like_ratio column storing the ratio between the number of likes and of dislikes


```python
cleaned['likes'] = pd.to_numeric(cleaned['likes'], errors='coerce').fillna(0)
cleaned['dislikes'] = pd.to_numeric(cleaned['dislikes'], errors='coerce').fillna(0)
cleaned['like_ratio'] = cleaned['likes'] / cleaned['dislikes'].replace({0: pd.NA})
```

### 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
cleaned['publish_time_parsed'] = pd.to_datetime(cleaned['publish_time'], errors='coerce')
cleaned['publish_interval_10min'] = cleaned['publish_time_parsed'].dt.floor('10min')
```

### 7. For each interval, determine the number of videos, average number of likes and of dislikes.


```python
interval_stats = cleaned.groupby('publish_interval_10min').agg(
    videos_count=('video_id','count'),
    avg_likes=('likes','mean'),
    avg_dislikes=('dislikes','mean')
).reset_index()
print("Step 7: Interval stats sample\n", interval_stats.head())
```

### 8. For each tag, determine the number of videos


```python
def split_tags(s):
    if pd.isna(s): return []
    return [t.strip().strip('"') for t in s.split('|') if t.strip()]

tags_df = cleaned[['video_id','country','tags','like_ratio']].copy()
tags_df['tags_list'] = tags_df['tags'].apply(split_tags)
tags_exploded = tags_df.explode('tags_list').rename(columns={'tags_list':'tag'})

tag_counts = tags_exploded.groupby('tag')['video_id'].count().reset_index(name='video_count')
print("Step 8: Tag counts sample\n", tag_counts.head())
```

### 9. Find the tags with the largest number of videos


```python
top_tags = tag_counts.sort_values('video_count', ascending=False).head(10)
print("Step 9: Top tags\n", top_tags)
```

### 10. For each (tag, country) pair, compute average ratio likes/dislikes


```python
tag_country_ratio = tags_exploded.groupby(['tag','country']).agg(
    avg_like_ratio=('like_ratio','mean'),
    videos_count=('video_id','count')
).reset_index().sort_values(['videos_count','avg_like_ratio'], ascending=[False, False])
print("Step 10: Tag-country ratio sample\n", tag_country_ratio.head())
```

### 11. For each (trending_date, country) pair, the video with the largest number of views


```python
top_per_trending_country = merged.sort_values(['country','trending_date','views'], ascending=[True,True,False]) \
                                 .groupby(['trending_date','country']).first().reset_index()
print("Step 11: Top video per trending_date-country\n", top_per_trending_country[['trending_date','country','video_id','views','title']].head())
```

```python
top_per_trending_country = merged.sort_values(['country','trending_date','views'], ascending=[True,True,False]) \
                                 .groupby(['trending_date','country']).first().reset_index()
print("Step 11: Top video per trending_date-country\n", top_per_trending_country[['trending_date','country','video_id','views','title']].head())
```

### 12. Divide trending_date into three columns: year, month, day


```python
def split_trending_date(s):
    try:
        yy, dd, mm = map(int, s.split('.'))
        yy += 2000 if yy < 100 else 0
        return pd.Series([yy, mm, dd])
    except:
        return pd.Series([pd.NA, pd.NA, pd.NA])

merged[['trending_year','trending_month','trending_day']] = merged['trending_date'].apply(split_trending_date)
print("Step 12: Split trending_date sample\n", merged[['trending_date','trending_year','trending_month','trending_day']].head())
```

### 13. For each (month, country) pair, the video with the largest number of views


```python
top_per_month_country = merged.sort_values(['trending_month','country','views'], ascending=[True,True,False]) \
                              .groupby(['trending_month','country']).first().reset_index()
print("Step 13: Top video per month-country\n", top_per_month_country[['trending_month','country','video_id','views','title']].head())
```

### 14. Read all json files with the video categories


```python
import glob, json

json_files = glob.glob("*.json")
categories_mapping = {}
for jf in json_files:
    with open(jf,'r', encoding='utf-8') as f:
        data = json.load(f)
    mapping = {str(item['id']): item['snippet']['title'] for item in data.get('items', []) if 'id' in item}
    country_code = Path(jf).stem.split('_')[0].upper()
    categories_mapping[country_code] = mapping
```

### 15. For each country, determine how many videos have a category that is not assignable.

```python
not_assignable_counts = []
for country in merged['country'].unique():
    df_c = merged[merged['country'] == country]
    mapping = categories_mapping.get(country, {})
    if mapping:
        missing = ~df_c['category_id'].astype(str).isin(mapping.keys())
        missing_count = missing.sum()
    else:
        missing_count = len(df_c)
    not_assignable_counts.append({
        'country': country,
        'total_videos': len(df_c),
        'not_assignable_count': missing_count
    })

not_assignable_df = pd.DataFrame(not_assignable_counts)
print("Step 15: Non-assignable categories per country\n", not_assignable_df)
```
